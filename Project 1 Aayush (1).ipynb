```python
import numpy as np
import pandas as pd
```

```python
housing=pd.read_csv('/content/Housing.csv')
```

```python
housing.head(25)
```

```python
housing['mainroad']=housing['mainroad'].apply({'yes':1,'no':0}.get)
housing['guestroom']=housing['guestroom'].apply({'yes':1,'no':0}.get)
housing['basement']=housing['basement'].apply({'yes':1,'no':0}.get)
housing['hotwaterheating']=housing['hotwaterheating'].apply({'yes':1,'no':0}.get)
housing['airconditioning']=housing['airconditioning'].apply({'yes':1,'no':0}.get)
housing['prefarea']=housing['prefarea'].apply({'yes':1,'no':0}.get)
housing['furnishingstatus']=housing['furnishingstatus'].apply({'furnished':1,'unfurnished':0,'semi-furnished':2}.get)







```

```python
housing.head()
```

```python
x=housing[['area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus'] ]
y=housing['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
y_pred=regression.predict(x_test)
```

```python
print(y_pred)
```

```python
regression.score(x,y)
```
